```python
import pandas as pd
import numpy as np

# First, recreate the original data dictionary
users = {'user_id': [100, 100, 101, 101, 102, 102, 103,103, 103, 104, 104, 105, 
                     106, 106, 106, 107, 107, 108, 108, 109],
         'DOB': [20000221, 20000221, 19901212, 19901212, 19831125, 19831125, 
                 19880920, 19880920, 19880920, 19750614,19750614, 19600522, 19950505, 
                 19950505, 19950505, 19770723, 19770723, 20010101, 20010101, 19990608],
         'movie_id': [1000, 1001, 1000, 1002, 1001,1002, 1000, 1002, 1002, 1001, 1002, 
                      1001, 1000, 1001, 1002, 1001, 1002, 1000, 1001, 1002],
         'timestamp': [20210728, 20210727, 20210726, 20210728, 20210727, 20210726, 20210728, 
                       20210727, 20210726,20210728, 20210727, 20210726, 20210728, 20210727, 20210726,
                       20210728, 20210727, 20210726, 20210727, 20210726],
         'rating': [9,8,7,5,8,7,9,10,5,6,8,8, 9, 8, 6, 8, 7, 8,6,8]}

# Create the DataFrame
Data2 = pd.DataFrame(users)

# Convert DOB to datetime
Data2['DOB'] = Data2['DOB'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))

# Calculate ages
Age2 = pd.to_datetime('today').year - Data2['DOB'].dt.year

# Add Age column
Data2.insert(2, "Age", Age2)

# Find min and max age
minAge = Data2['Age'].min()
maxAge = Data2['Age'].max()

# Create bins and labels
bins = np.linspace(minAge, maxAge, 4)
labels = ['young', 'middle age', 'senior']

# Apply binning
Data2['bins'] = pd.cut(Data2['Age'], bins=bins, labels=labels, include_lowest=True)

# Now perform the groupby and aggregation
result = Data2.groupby('bins').agg({'rating':['min','max']})

# Display the result
print("Minimum and maximum ratings per age group (bins):")
display(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/YEO/OneDrive/Desktop/Degree/Year 2/SEM 3 LONG SEM (2025)/TDW6323-DATA WRANGLING AND VISUAL (DW1,1C)/LAB/LAB 9/AirQualityUCI.csv", header=[0], sep=";")

# Handle missing data
threshold = data.isnull().sum() / len(data) 
remove_col = threshold[threshold > 0.8].index  # get columns with >80% missing data
data.drop(remove_col, axis=1, inplace=True)

threshold = data.isnull().sum(axis=1) / (len(data.columns) - 1)
remove_row = threshold[threshold > 0.8].index  # get rows with >80% missing data
data.drop(remove_row, axis=0, inplace=True)

print("Data after removing high missing value columns/rows:")
print(data.head())
print(f"Data length: {len(data)}")

# Convert date and time columns
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data['Time'] = pd.to_datetime(data['Time'], format='%H.%M.%S').dt.time

# Create a datetime column by combining Date and Time
data['DateTime'] = pd.to_datetime(data['Date'].astype(str) + ' ' + data['Time'].astype(str))

# Set DateTime as index for resampling
data.set_index('DateTime', inplace=True)

# 2. Resample the data by producing average value by week (downsample)
# Select only numeric columns for resampling
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Resample to weekly averages
weekly_data = data[numeric_columns].resample('W').mean()

print("\nWeekly resampled data:")
print(weekly_data.head())
print(f"Original data points: {len(data)}")
print(f"Weekly averaged data points: {len(weekly_data)}")

# Optional: Handle any remaining NaN values in the resampled data
weekly_data_clean = weekly_data.dropna()
print(f"Clean weekly data points (after dropping NaN): {len(weekly_data_clean)}")

# Example visualization of one pollutant over time
if 'CO(GT)' in weekly_data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(weekly_data.index, weekly_data['CO(GT)'])
    plt.title('Weekly Average CO Concentration Over Time')
    plt.xlabel('Date')
    plt.ylabel('CO Concentration')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Display basic statistics of the weekly data
print("\nWeekly data statistics:")
print(weekly_data.describe())
```

8b.4 Wrapper Method – Backward Elimination



```python
#1.	Use linear regression to perform backward elimination. Which features are selected?
cols=list(x.columns) #get the columns
while(len(cols)>0):  #when there are columns available in cols
    p=[]
    x_1=x[cols]
    x_1=sm.add_constant(x_1)
    model=sm.OLS(y,x_1).fit() 
#sm is an object of stat models, and we use OLS froms statmodels library
    p=pd.Series(model.pvalues.values[1:], index=cols)
    pmax=max(p)
    features_with_p_max=p.idxmax()
    if(pmax>0.05):  #p is the null hypothesis p value, if greater than 0.05, remove feature
        cols.remove(features_with_p_max)
    else:
        break
    summary = model.summary()
print(summary)
selected_features=cols
print(selected_features)

```
