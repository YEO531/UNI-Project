```python
#Union of Two Fuzzy Sets
C = dict()
D = dict()
Z = dict()
 
C = {"a": 0.1, "b": 0.4, "c": 0.7, "d": 0.9}
D = {"a": 0.4, "b": 0.8, "c": 0.2, "d": 0.6}
 
print('The First Fuzzy Set is :', C)
print('The Second Fuzzy Set is :', D)
 
 
for C_key, D_key in zip(C, D):
    C_value = C[C_key]
    D_value = D[D_key]
 
    if C_value > D_value:
        Z[C_key] = C_value
    else:
        Z[D_key] = D_value
         
print('Fuzzy Set Union is :', Z)
```

```python
# Intersection of Two Fuzzy Sets
C = dict()
D = dict()
Z = dict()
 
C = {"a": 0.1, "b": 0.4, "c": 0.7, "d": 0.9}
D = {"a": 0.4, "b": 0.8, "c": 0.2, "d": 0.6}
 
print('The First Fuzzy Set is :', C)
print('The Second Fuzzy Set is :', D)
 
 
for C_key, D_key in zip(C, D):
    C_value = C[C_key]
    D_value = D[D_key]
 
    if C_value < D_value:
        Z[C_key] = C_value
    else:
        Z[D_key] = D_value
         
print('Fuzzy Set Intersection is :', Z)
```

```python
#Complement
C = dict()
Z = dict()
 
C = {"a": 0.1, "b": 0.4, "c": 0.6, "d": 0.3}
print('The Fuzzy Set is :', C) 
 
for C_key in C:
   Z[C_key]= 1 - C[C_key]
         
print('Fuzzy Set Complement is :', Z)
```

```python
#Difference between Two Fuzzy Sets
C = dict()
D = dict()
Z = dict()
 
C = {"a": 0.1, "b": 0.4, "c": 0.7, "d": 0.9}
D = {"a": 0.4, "b": 0.8, "c": 0.2, "d": 0.6}
 
print('The First Fuzzy Set is :', C)
print('The Second Fuzzy Set is :', D)

for C_key, D_key in zip(C, D):
    C_value = C[C_key]
    D_value = D[D_key]
    D_value = 1 - D_value
 
    if C_value < D_value:
        Z[C_key] = C_value
    else:
        Z[D_key] = D_value
         
print('Fuzzy Set Difference is :', Z)
```
