# Section F: Data Collections
Namely list, tuple, set and dictionary

## 1. List
List is mutable and it is an ordered

```python
# Creating List
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
records = ['Jackson', 10, ['TES6313', 'A+']]

# Adding item. It is mutable
numbers.insert(3,'ABC')
numbers.append(200) 
print(numbers)

# Replace item. It is mutable
numbers[1] = 50
print(numbers)

# Remove item.  It is mutable
numbers.remove(1)
print(numbers)

del numbers[2]  
print(numbers)

numbers.pop(1)
print(numbers)

# Combine lists
print(numbers + records) 

# Display nested list. It is an ordered.
print(records[2][0])
```

## 2. Tuple
Tuple is immutable but it is an ordered

```python
# It is an ordered
coordinate_point = (10, 30)
print(coordinate_point[0])

# It is immutable
#coordinate_point[0]=20
```

# Exercises

1. Create an input and output for a simple expert system GUI.
2. Based on question 1, the output is getting from a text file.

```python
# Question 1
import tkinter as tk
from tkinter import messagebox

# Predefined expert system responses
responses = {
    "happy": "Take her out for a treat!",
    "sad": "Give her some chocolate and listen to her.",
    "angry": "Let her cool down, then talk to her.",
    "bad mood": "Send her to karaoke!"
}

def on_button_click():
    mood = entry.get().strip().lower()  # Get user input and convert to lowercase
    response = responses.get(mood, "Sorry, I don't have advice for this mood.")  # Get response or default
    messagebox.showinfo("Message", response)

# Set up GUI
window = tk.Tk()
window.title("Roommate Diagnose Expert System")
window.geometry("400x200+500+300")

# Label
label = tk.Label(window, text="Please enter your roommate's mood:")
label.pack(pady=10)

# Input field
entry = tk.Entry(window, width=40)
entry.insert(0, "bad mood")
entry.pack(pady=12)

# Button
button = tk.Button(window, text="Ok", command=on_button_click)
button.pack(pady=12)

# Start GUI loop
window.mainloop()
```

```python
# Question 2
import tkinter as tk
from tkinter import messagebox

def on_button_click():
    # messagebox.showinfo("Message",entry.get())
    # messagebox.showinfo("Message","Send her to karaoke!")
    mood = entry.get().strip()  # Get user input and strip whitespace
    try:
        with open("responses.txt", "r") as file:  # Read responses from the text file
            for line in file:
                mood_response = line.strip().split(":", 1)  # Split into mood and response
                if len(mood_response) == 2 and mood_response[0].lower() == mood.lower():
                    messagebox.showinfo("Message", mood_response[1])
                    return
        # If no match is found
        messagebox.showinfo("Message", "Sorry, no advice available for this mood.")
    except FileNotFoundError:
        messagebox.showerror("Error", "Response file not found. Please create 'responses.txt'.")

window = tk.Tk()
window.title("Roommate Diagnose Expert System")
window.geometry("300x200+500+300")
    
label = tk.Label(window, text="Please enter your roommate's mood")
label.pack(pady=10)

entry = tk.Entry(window, width=40)
entry.insert(0,"bad mood")
entry.pack(pady=12)

button = tk.Button(window, text="Ok", command=on_button_click)
button.pack(pady=12)

window.mainloop()
```
